## Boxplot
### Veri seti hikayesi

##### total_bill : yemeğin toplam fiyatı (bahşiş ve vergi dahil)
##### tip ? bahşiş
##### sex : ücreti ödeyen kişinin cinsiyeti(0=male, 1=female)
##### grupta sigara içen var mı? (0= No, 1=Yes) 
##### day : gün(3 = Thur, 4=Fri, 5=Sat, 6=Sun)
##### ne zaman? (0=Day, 1=Night) 
##### grupta kaç kişi var?

```python
import seaborn as sns 
import numpy as np
import pandas as pd
tips = sns.load_dataset("tips")
df = tips.copy()
df.head()
```

```python
df.describe().T
```

```python
df["sex"].value_counts()
```

```python
df["smoker"].value_counts()
```

```python
df["day"].value_counts()
```

```python
df["time"].value_counts()
```

```python
# Boxplot
```

```python
sns.boxplot(x = df["total_bill"]);
```

```python
sns.boxplot(x=df["total_bill"], orient="x")
```

```python
# crossovers
```

```python
df.describe().T
```

```python
# Which days do we earn more?
```

```python
sns.boxplot(x="day", y="total_bill", data=df);
```

```python
# day or night ?
```

```python
sns.boxplot(x="time", y="total_bill", data=df);
```

```python
# number of people - earnings 
```

```python
sns.boxplot(x="size", y="total_bill", data=df);
```

```python
sns.boxplot(x="day", y="total_bill", hue="sex", data=df);
```

```python
# Violin
```

```python
df.head()
```

```python
sns.catplot(y= "total_bill", kind = "violin", data=df);
```

```python
# crossovers
```

```python
sns.catplot(x="day", y="total_bill",kind = "violin" ,data=df);
```

```python
sns.catplot(x="day", y="total_bill",hue="sex", kind = "violin" ,data=df);
```

```python
# correlation graphs 
# scatterplot
```

```python
sns.scatterplot(x = "total_bill", y= "tip", data=df)
```

```python
# correlation crossovers
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="time", data=df);
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="time",style="time" ,data=df);
```

```python
sns.scatterplot(x="total_bill", y="tip", hue="day",style="time" ,data=df);
```

```python
sns.scatterplot(x="total_bill", y="tip",hue="size" ,size="size" ,data=df);
```
